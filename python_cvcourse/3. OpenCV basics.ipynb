```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("Computer-Vision-with-Python/DATA/00-puppy.jpg")
```

```python
type(img)  # it is "NoneType" if image was not found
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# MATPLOTLIB: RGB
# OPENCV: BGR
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread("Computer-Vision-with-Python/DATA/00-puppy.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (1000, 400))
new_img.shape
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.8
h_ratio = 0.2
```

```python
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
# flipping images (0: x axis, 1: y axis, -1: both axis)
new_img = cv2.flip(fix_img, -1)
plt.imshow(new_img)
```
